```python
df=df.dropna(axis=1,thresh=len(df)*0.1)
```

```python
df.select_dtypes(include=np.number).columns
```

```python
df[df.select_dtypes(include=np.number).columns]=df.select_dtypes(include=np.number).fillna(df.select_dtypes(include=np.number).mean())
```

```python
df.isnull().sum().sort_values(ascending=False).head(20)
```

```python
df=df.dropna(axis=1,how="all")
```

```python
df["Shelf"]=df["Shelf"].fillna("unkonwn")
```

```python
df.isnull().sum().sort_values(ascending=False).head(20)
```

```python
df["Billing Address Line 2"]=df["Billing Address Line 2"].fillna("not provided")
df["Shipping Address Line 2"]=df["Shipping Address Line 2"].fillna("not provided")
```

```python
bundles_cols=[
    "Combination Identifier",
    "Combination Description",
   "Bundle SKU Code Number"]
for col in bundles_cols:
    df[col]= df[col].fillna("not bundle")
    
```

```python
shipping_cols=[
    "Delivery Time", 
"Shipping Courier" ,              
"Shipping Tracking Status",          
"Shipping Courier Status",
"Shipping provider",
"Tracking Number"]
for col in shipping_cols:
    df[col]=df[col].fillna("not shipped")
```

```python
ops_cols = ["Dispatch Date",
"Batch Code",
"Vendor Batch Number",
"GSTIN",
"Invoice Code",
"Picklist Code",
"Invoice Created"]
for col in ops_cols:
    df[col]=df[col].fillna("not genrated")
```

```python
df['tags']=df['tags'].fillna("no tag")
```

```python
df.fillna(0,inplace=True)
```
